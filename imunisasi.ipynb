```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfBCG=pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv',na_values = ['n.a'])
dfBCG=dfBCG.interpolate()

dfCampak=pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv',na_values = ['n.a'])
dfCampak=dfCampak.interpolate()

dfDPT=pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv',na_values = ['n.a'])
dfDPT=dfDPT.interpolate()

dfPolio=pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv',na_values = ['n.a'])
dfPolio=dfPolio.interpolate()
```

```python
plt.subplot(221)
plt.bar(dfBCG['Tahun'],dfBCG['% Balita yang pernah mendapat imunisasi BCG'],color='red')
plt.title('BCG')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,110,20))

plt.subplot(222)
plt.bar(dfCampak['Tahun'],dfCampak['% Balita yang pernah mendapat imunisasi Campak'],color='green')
plt.title('Campak')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,110,20))

plt.subplot(223)
plt.bar(dfDPT['Tahun'],dfDPT['% Balita yang pernah mendapat imunisasi DPT'],color='yellow')
plt.title('DPT')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,110,20))

plt.subplot(224)
plt.bar(dfPolio['Tahun'],dfPolio['% Balita yang pernah mendapat imunisasi Polio'],color='blue')
plt.title('Polio')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,110,20))

plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.suptitle('Persentase balita terimunisasi 1995-2017')
```

```python
bcgMinus=100-dfBCG['% Balita yang pernah mendapat imunisasi BCG']
campakMinus=100-dfCampak['% Balita yang pernah mendapat imunisasi Campak']
dptMinus=100-dfDPT['% Balita yang pernah mendapat imunisasi DPT']
polioMinus=100-dfPolio['% Balita yang pernah mendapat imunisasi Polio']
```

```python
plt.subplot(221)
plt.bar(dfBCG['Tahun'],bcgMinus,color='red')
plt.title('BCG')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,60,20))

plt.subplot(222)
plt.bar(dfCampak['Tahun'],campakMinus,color='green')
plt.title('Campak')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,60,20))

plt.subplot(223)
plt.bar(dfDPT['Tahun'],dptMinus,color='yellow')
plt.title('DPT')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,60,20))

plt.subplot(224)
plt.bar(dfPolio['Tahun'],polioMinus,color='blue')
plt.title('Polio')
plt.xticks(range(1995,2017),rotation=90,size=5)
plt.yticks(range(0,60,20))

plt.subplots_adjust(wspace=0.2,hspace=0.5)
plt.suptitle('Persentase balita tidak terimunisasi 1995-2017')
```
